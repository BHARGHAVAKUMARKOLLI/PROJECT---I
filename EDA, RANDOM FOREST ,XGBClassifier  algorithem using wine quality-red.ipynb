```python
#Categorizing Data
```

```python
wine.quality = [1 if each >= 7 else 0 for each in wine.quality]
wine.quality
```

```python
wine.quality.value_counts()
```

```python
y = wine.quality.values
print('wine quality values is',y)
```

```python
x_data = wine.drop(["quality"],axis=1)
print('drop the quality', x_data)
```

```python
#DATA NORMALIZATION
```

```python
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
print('x value is ', x)
```

```python
##Spliting data or Creating Train and test data files
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#Random Forest Classifier

##Random forestıclassifier creates a set of decision trees from randomly selected subset of training set.
##It then aggregates the votes from different decision trees to decide the final class of the test object.
##First, the Random Forest algorithm is a controlled classification algorithm. (Supervised classification algorithm).
##As we can see from the name, the algorithm simply creates a forest randomly. There is a direct relationship between the number of trees in the algorithm and the results they can achieve. As the number of trees increases, 
##we get a definite result.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score 
```

```python
RF = RandomForestClassifier(n_estimators=200, random_state=1)
RF.fit(train_x,train_y)
```

```python
### PREDICT THE MODEL
predictions = RF.predict(test_x)
```

```python
score = round(accuracy_score(test_y, predictions), 5)
```

```python
print("Random Forest Score is  {}".format(score))
```

```python
##CROSS VALIDATION
from sklearn.model_selection import cross_val_score, train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
##Model Prediction XGBOOST
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(train_x,train_y)## fitting the model by using XGBClassifier
```

```python
pred = model.predict(test_x) ## predicting the model (XGBClassifier)
```

```python
##To find accuracy of XGBClassifier
score = accuracy_score(test_y,pred)
print('Accuracy of XGBClassifier is', score)
```
